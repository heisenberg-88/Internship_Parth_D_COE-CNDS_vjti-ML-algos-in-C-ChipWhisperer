```python
# creating the column names array
namesdata = []
i = 0
while(True):
    if(i==5001):
        break
    namesdata.append(str(i))
    i+=1
```

```python
df = pd.read_csv('final_dataset_randomized.csv',names=namesdata)
# this will put namesdata in the place of column names
```

```python
# // only run single time
feature_column_names = []
namesdata.pop()
feature_column_names = namesdata
feature_column_names[-1]
```

```python
features = df[feature_column_names]
```

```python
labels = df['5000']
```

```python
feature_array = features.to_numpy()
feature_array.shape
```

```python
def isfloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False

# 0.473634.14
# lot of values in the dataset are like 0.00345628.14  which is invalid and the garbage values 
# after second '.' is due to csv format. so , i created this function to return the index upto which the float number is valid
def floatmaker(data):
    data = str(data)
    first_dot = False
    second_dot = False
    lastindex = 0
    i = 0
    for c in data:
        if(second_dot==True):
            break
        if(c=='.' and first_dot==False):
            first_dot=True
        elif(c=='.' and first_dot==True):
            second_dot = True
            lastindex = i
        i+=1
    return data[0:lastindex]

```

```python
cleaned_features_array = []
for curr_list in feature_array:
    mylist = []
    for i in curr_list:
        if(str(i)=='.' or str(i)=='-'):
            mylist.append(float(0))
        elif(isfloat(i)==False):
            print(i+" is not float")
            newfloat = floatmaker(i)
            mylist.append(float(newfloat))
            print("old "+i)
            print("new "+newfloat)
            print("--------")
        else:
            mylist.append(float(i))
         
    
    cleaned_features_array.append(mylist)
```

```python
import numpy as np
feature_array_final = np.array([np.array(xi) for xi in cleaned_features_array])
feature_array_final.shape
```

```python
label_array = labels.to_numpy()
```

```python
label_array.shape
```

```python
feature_array_final[0].shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dropout
```

```python
# creating the LSTM network
model = Sequential()
model.add(Dense(256,activation='relu',input_shape=(5000,1)))
model.add(LSTM(128, activation='tanh',recurrent_activation='sigmoid'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(6,activation='softmax'))

print(model.summary())
```

```python
# compile and fit
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
modelsaved = model.fit(feature_array_final, label_array, epochs=10,validation_split=0.1)
```

```python
# make predictions
trainPredict = model.predict(feature_array_final)
```

```python
trainPredict[0]
```

```python
loss_train = modelsaved.history['loss']
loss_val = modelsaved.history['val_loss']
```
