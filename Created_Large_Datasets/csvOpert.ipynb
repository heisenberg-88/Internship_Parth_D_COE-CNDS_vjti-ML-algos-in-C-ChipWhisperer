```python
filenames = ['DECT.csv','knn.csv','LIR.csv','LOR.csv','NB.csv','SVM.csv']
```

```python
i = 0
for file in filenames:
    olddf = pd.read_csv(file)
    mylist = [i]*5000
    labels = pd.DataFrame(mylist)
    label_and_data = pd.concat([olddf,labels],axis=1)
    label_and_data.to_csv('_final_'+file , index = False)
    i+=1
    
```
